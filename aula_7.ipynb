```python
def calculadora_simples():
    # Inicia um bloco try-except para lidar com possíveis erros durante a execução
    try:
        # Solicita os números ao usuário
        num1_str = input("Digite o primeiro número: ")
        num2_str = input("Digite o segundo número: ")

        # Converte para float (aceita decimais)
        num1 = float(num1_str)
        num2 = float(num2_str)

        # Solicita a operação desejada
        operacao = input("Digite a operação (soma, subtracao, multiplicacao, divisao): ").lower()

        # Verifica qual operação foi escolhida
        if operacao == 'soma':
            resultado = num1 + num2
            print(f"Resultado da soma: {resultado}")
        elif operacao == 'subtracao':
            resultado = num1 - num2
            print(f"Resultado da subtração: {resultado}")
        elif operacao == 'multiplicacao':
            resultado = num1 * num2
            print(f"Resultado da multiplicação: {resultado}")
        elif operacao == 'divisao':
            # Tratamento especial para divisão por zero
            if num2 == 0:
                print("Erro: divisão por zero não é permitida.")
            else:
                resultado = num1 / num2
                print(f"Resultado da divisão: {resultado}")
        else:
            print("Erro: Operação inválida. Use 'soma', 'subtracao', 'multiplicacao' ou 'divisao'.")

    except ValueError:
        # Caso o usuário digite algo que não seja número
        print("Erro: Entrada inválida. Por favor, digite apenas números.")
    except Exception as e:
        # Captura qualquer outro erro inesperado
        print(f"Ocorreu um erro inesperado: {e}")

# Chama a função para iniciar
calculadora_simples()
```

```python
import requests
import pandas as pd
import json

def consulta_cep(cep):
    """
    Consulta informações de endereço a partir de um CEP usando a API ViaCEP.
    Retorna um dicionário com os dados do CEP ou None em caso de erro.
    """
    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        if response.status_code == 200:
            dados = response.json()
            if "erro" in dados:
                print("CEP não encontrado.")
                return None
            else:
                print(f"CEP: {dados.get('cep')}")
                print(f"Logradouro: {dados.get('logradouro')}")
                print(f"Bairro: {dados.get('bairro')}")
                print(f"Cidade: {dados.get('localidade')}")
                print(f"Estado: {dados.get('uf')}")
                return dados
        else:
            print("Erro na requisição. Código:", response.status_code)
            return None
    except Exception as e:
        print("Erro ao consultar CEP:", e)
        return None

# ============================================
# Exemplo de uso e salvamento em JSON
# ============================================
cep_para_salvar = input("Digite o CEP para salvar em JSON (somente números): ")
dados_cep = consulta_cep(cep_para_salvar)

if dados_cep:
    nome_arquivo_json = f"cep_{cep_para_salvar}.json"
    with open(nome_arquivo_json, 'w', encoding='utf-8') as f:
        json.dump(dados_cep, f, ensure_ascii=False, indent=4)
    print(f"Dados do CEP salvos em {nome_arquivo_json}")
else:
    print("Não foi possível obter dados para salvar em JSON.")

# ============================================
# Exemplo de uso e salvamento em CSV
# ============================================

cep_para_salvar_csv = input("Digite o CEP para salvar em CSV (somente números): ")
dados_cep_csv = consulta_cep(cep_para_salvar_csv)

if dados_cep_csv:
    df = pd.DataFrame([dados_cep_csv])
    nome_arquivo_csv = f"cep_{cep_para_salvar_csv}.csv"
    df.to_csv(nome_arquivo_csv, index=False, encoding='utf-8')
    print(f"Dados do CEP salvos em {nome_arquivo_csv}")
else:
    print("Não foi possível obter dados para salvar em CSV.")
```

```python
import pandas as pd

def validar_cpf(cpf: str) -> bool:
    """
    Valida um CPF (Cadastro de Pessoa Física).
    Retorna True se o CPF for válido, False caso contrário.
    """

    # Remove caracteres não numéricos
    cpf = ''.join(filter(str.isdigit, cpf))

    # Verifica se tem 11 dígitos
    if len(cpf) != 11:
        return False

    # Elimina CPFs com todos os dígitos iguais (ex: 11111111111)
    if cpf == cpf[0] * 11:
        return False

    # Cálculo do primeiro dígito verificador
    soma = 0
    for i in range(9):
        soma += int(cpf[i]) * (10 - i)
    digito1 = (soma * 10) % 11
    digito1 = 0 if digito1 == 10 else digito1

    # Cálculo do segundo dígito verificador
    soma = 0
    for i in range(10):
        soma += int(cpf[i]) * (11 - i)
    digito2 = (soma * 10) % 11
    digito2 = 0 if digito2 == 10 else digito2

    # Verifica se os dígitos calculados batem com os informados
    return digito1 == int(cpf[9]) and digito2 == int(cpf[10])


cpfs_para_salvar = []

print("Digite os CPFs que deseja validar e salvar (digite 'fim' para terminar):")
while True:
    cpf_input = input("Digite um CPF: ")
    if cpf_input.lower() == 'fim':
        break

    is_valid = validar_cpf(cpf_input)
    cpfs_para_salvar.append({
        'CPF': cpf_input,
        'Valido': is_valid
    })

if cpfs_para_salvar:
    df_cpfs = pd.DataFrame(cpfs_para_salvar)
    nome_arquivo_csv = 'cpfs_validados.csv'
    df_cpfs.to_csv(nome_arquivo_csv, index=False, encoding='utf-8')
    print(f"\n{len(cpfs_para_salvar)} CPFs validados e salvos em '{nome_arquivo_csv}'.")
    display(df_cpfs.head())
else:
    print("Nenhum CPF foi inserido para salvar.")
```

```python
def calcular(operacao: str, x: float, y: float) -> float:
    if operacao == '+':
        return x + y
    elif operacao == '-':
        return x - y
    elif operacao == '*':
        return x * y
    elif operacao == '/':
        if y == 0:
            raise ValueError("Divisão por zero não permitida.")
        return x / y
    elif operacao == '**':
        return x ** y
    elif operacao == '%':
        return x % y
    else:
        raise ValueError("Operação inválida.")

print("=== Calculadora Otimizada ===")
print("Operações disponíveis: +  -  *  /  **  %")
print("Digite 'sair' para encerrar.\n")

while True:
    operacao = input("Escolha a operação: ").strip()
    if operacao.lower() == 'sair':
        print("Encerrando calculadora. Até mais!")
        break

    try:
        x = float(input("Digite o primeiro número: "))
        y = float(input("Digite o segundo número: "))
        resultado = calcular(operacao, x, y)
        print(f"Resultado: {resultado}\n")
    except Exception as e:
        print(f"Erro: {e}\n")
```
